```python
from google.colab import drive
import os
from pydub import AudioSegment
import speech_recognition as sr

# Mount Google Drive
drive.mount('/content/drive')

# Define the directory path
directory = "/content/drive/Shared drives/Projects/Teneika Data Analyst Program/Webscraping Conference Schedules/"

# Initialize the recognizer
recognizer = sr.Recognizer()

# Define the input and output file paths
audio_file_path = os.path.join(directory, "GMT20241207-152550_Recording.m4a")
wav_file_path = os.path.join(directory, "Recording.wav")

# Convert the file to WAV format
audio_file = AudioSegment.from_file(audio_file_path)
audio_file.export(wav_file_path, format="wav")

# Load the audio file
audio = AudioSegment.from_wav(wav_file_path)

# Start at the 5-minute mark (5 minutes in milliseconds)
# start_time = 5 * 60 * 1000  # 5 minutes in milliseconds
start_time = (5 * 60 + 15) * 1000  # 5 minutes and 15 seconds in milliseconds
audio = audio[start_time:]  # Slice audio from 5-minute mark to the end

# Set chunk size (30 seconds in milliseconds)
chunk_size = 50 * 1000  # 50 seconds
chunks = [audio[i:i + chunk_size] for i in range(0, len(audio), chunk_size)]


# Transcribe each chunk
transcriptions = []
for index, chunk in enumerate(chunks):
    chunk_file_path = os.path.join(directory, f"chunk_{index}.wav")
    chunk.export(chunk_file_path, format="wav")  # Export the chunk as a WAV file

    with sr.AudioFile(chunk_file_path) as source:
        try:
            recognizer.adjust_for_ambient_noise(source)
            audio_content = recognizer.record(source)
            text = recognizer.recognize_google(audio_content)
            transcriptions.append(text)
            print(f"Chunk {index + 1} transcribed successfully.")
        except sr.RequestError as e:
            print(f"RequestError for chunk {index + 1}: {e}")
        except sr.UnknownValueError:
            print(f"Chunk {index + 1}: Audio not clear.")
        except Exception as e:
            print(f"Unexpected error for chunk {index + 1}: {e}")

# Combine all transcriptions
full_transcription = " ".join(transcriptions)

# Save the transcription to a file
output_file_path = os.path.join(directory, "1207transcription.txt")
with open(output_file_path, "w") as f:
    f.write(full_transcription)

# Display a preview of the transcription
print("Transcription Preview:")
print(full_transcription[:1000])  # First 1000 characters
```
