## Part 1: Introduction to Pandas
Pandas is a powerful library for data manipulation and analysis. Below are some common tasks:
- Loading data from a CSV file.
- Inspecting the data (head, tail, info).
- Cleaning and manipulating data.

---

```python
# Part 1: Pandas Basics
import pandas as pd

# Load a sample dataset
data = {
    "Product": ["A", "B", "C", "D"],
    "Sales": [200, 150, 300, 400],
    "Region": ["North", "South", "East", "West"]
}
df = pd.DataFrame(data)

# Display the DataFrame
print(df)

# Basic Operations
print(df.head())  # Display the first 5 rows
print(df.info())  # Get information about the DataFrame
print(df.describe())  # Summary statistics
```


## Part 2: Introduction to Matplotlib and Seaborn
Matplotlib and Seaborn are popular libraries for creating visualizations. Below are some examples:
- Line plots, bar plots, and scatter plots with Matplotlib.
- Advanced plots with Seaborn (heatmaps, pairplots, etc.).

---

```python
# Part 2: Visualization with Matplotlib
import matplotlib.pyplot as plt

# Bar Chart
plt.bar(df["Product"], df["Sales"])
plt.xlabel("Product")
plt.ylabel("Sales")
plt.title("Sales by Product")
plt.show()
```

```python
# Part 2: Visualization with Seaborn
import seaborn as sns

# # Create a sample heatmap
# sns.heatmap(df.corr(), annot=True, cmap="coolwarm")
# plt.title("Correlation Heatmap")
# plt.show()
# Select numeric columns for correlation
numeric_df = df.select_dtypes(include="number")

# Create a heatmap
sns.heatmap(numeric_df.corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

## Dataset:

```python
# Simulated Dataset
sales_data = {
    "Product": ["A", "B", "C", "A", "B", "C", "A", "B", "C"],
    "Sales": [200, 150, 300, 250, 100, 350, 300, 200, 400],
    "Region": ["North", "South", "East", "North", "South", "East", "West", "North", "South"]
}
sales_df = pd.DataFrame(sales_data)
sales_df.to_csv("sales_data.csv", index=False)

print("Dataset saved as 'sales_data.csv'")
```

```python
# Lab Solution
# Load the dataset
df = pd.read_csv("sales_data.csv")

# Total sales by region
total_sales_by_region = df.groupby("Region")["Sales"].sum()
print(total_sales_by_region)

# Product with the highest sales
highest_sales_product = df.groupby("Product")["Sales"].sum().idxmax()
print(f"Product with the highest sales: {highest_sales_product}")

# Visualizations
# Bar chart for total sales by region
total_sales_by_region.plot(kind="bar", title="Total Sales by Region")
plt.ylabel("Total Sales")
plt.show()

# Pie chart for sales distribution of products
df.groupby("Product")["Sales"].sum().plot(kind="pie", autopct="%1.1f%%", title="Sales Distribution by Product")
plt.ylabel("")  # Remove the default ylabel
plt.show()
```
